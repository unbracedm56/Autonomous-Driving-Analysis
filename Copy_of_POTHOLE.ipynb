```python
import kagglehub

# Download latest version of the dataset
path = kagglehub.dataset_download("anggadwisunarto/potholes-detection-yolov8")

print("Path to dataset files:", path)
!ls -R $path | head -50  # show first 50 lines of dataset structure
```

```python
import os

for root, dirs, files in os.walk(path):
    level = root.replace(path, "").count(os.sep)
    indent = " " * 2 * level
    print(f"{indent}{os.path.basename(root)}/")
    subindent = " " * 2 * (level + 1)
    for f in files:
        print(f"{subindent}{f}")
```

```python
import yaml

yaml_path = os.path.join(path, "data.yaml")

# Load YAML
with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

# Fix train/val paths to Colab absolute paths
data["train"] = os.path.join(path, "train/images")
data["val"]   = os.path.join(path, "valid/images")

# Save corrected YAML
with open(yaml_path, "w") as f:
    yaml.dump(data, f)

print("✅ data.yaml fixed")
print(data)
```

```python
from ultralytics import YOLO

# Load pretrained YOLOv8 small model
model = YOLO("yolov8s.pt")

# Train
model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    name="pothole_detector"
)
```

```python
print(model.val())
```

```python
results = model.predict(source=os.path.join(path, "valid/images"), save=True, conf=0.5)
```
